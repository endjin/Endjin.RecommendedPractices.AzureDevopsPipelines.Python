```python
import pandas as pd
import numpy as np
from pathlib import Path
import seaborn as sns

from worldbank_data import DataWrangler
```

```python
data_wrangler = DataWrangler()
source_spreadsheet = '68fd98a4-a721-4a4b-ba8b-4fa738699aa2_Data.csv'
mapping_filename = 'country_metadata.csv'
df = data_wrangler.load_raw_data_from_excel(source_spreadsheet)
df.head(20)
```

```python
df = data_wrangler.unpivot_data(df)
df.head(20)
```

```python
df = data_wrangler.clean_years(df)
df
```

```python
df = data_wrangler.assign_nan_empty_metrics(df)
# df['Metric'] = df['Metric'].replace('..', np.nan)
```

```python
# df.loc[(df['Country Name']== 'American Samoa') & (df['Series Name'] == 'Population, female')]
df.loc[(df['Country Name']== 'Afghanistan') & (df['Series Name'] == 'GDP (current US$)')].head(50)
```

```python
df = data_wrangler.fill_empty_cells(df)
# df['Metric'] = df.groupby(['Country Name', 'Series Name'])['Metric'].ffill()
df.loc[(df['Country Name']== 'Afghanistan') & (df['Series Name'] == 'GDP (current US$)')].head(50)
```

```python
df = data_wrangler.pivot_data(df)
df
```

```python
df = data_wrangler.load_and_apply_mapping(df, mapping_filename)
df
```

```python
df.loc[df['Region'].isna(), 'Country Name'].value_counts()
```

```python
df = df.loc[~df['Region'].isna(), :]
df
```

```python
df = data_wrangler.convert_to_float(df, 'GDP per capita (current US$)')
df
```

```python
df = data_wrangler.data_to_log(df, 'GDP per capita (current US$)')
df
```

```python
df_2020 = df.loc[df['Year'] == '2020']
df_2020
```

```python
sns.scatterplot(x="GDP per capita (current US$)", y="Life expectancy at birth, total (years)", hue="Region", 
            sizes=(40, 400), alpha=.5, palette="muted", data=df_2020)
```

```python
df.to_csv('data.csv', index=False)
```
